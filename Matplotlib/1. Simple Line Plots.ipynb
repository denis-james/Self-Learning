# Simple Line Plots

The simplest to visualize out of all the functions is the $y=f(x)$. Let's create a simple plot of this type.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-v0_8-whitegrid')
```

For all Matplotlib Plots, we start of by creating an instance of Figure and an Axes

```python
fig=plt.figure()
ax=plt.axes()
```

In Matplotlib, the *figure* (an instance of the class `plt.Figure`) can be thought of as a single generated image instance that contains all the objects representing the different visualizations such as axes, graphs, text, and labels. The *axes* (an instance of the class `plt.Axes)` is what we see above: a bounding grid with lines and labels, which will eventually contain the plot elements that make up our visualization. Throughout this book, we'll commonly use the variable name `fig` to refer to a figure instance, and `ax` to refer to an axes instance or group of axes instances.

After defining the axes we can plot our graph in that given axes.

```python
fig=plt.figure()
ax=plt.axes()
x=np.linspace(-10,10,201,)

ax.plot(x,np.sin(np.pi*x))
```

Or we can use the Matlab interface and let the figure and axes be created for us in the background.

```python
plt.plot(x,np.sin(np.pi*x))
```

Say we want to graphs multiple equations on the same axes, we can simply call the plot function again

```python
plt.plot(x,np.sin(np.pi*x))
plt.plot(x,np.sin(np.pi*x-np.pi/2))
plt.plot(x,np.cos(np.pi*x))
```

Or we can use the same ax instance to plot another graph in the Object-Oriented approach. Remember, to define the axes before plotting.

```python
ax=plt.axes()
print(x)
ax.plot(x,np.sin(np.pi*x))
ax.plot(x,np.sin(np.pi*x-np.pi/2))
ax.plot(x,np.cos(np.pi*x))
```

# Adjusting the Plot: Line Colors and Styles

The first adjustment one would like to make can be the color of the plot. This can be done on the `plt.plot()` function by specifying the `color` argument withing the function. This function can take multiple formats as an argument for this function.

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

If no color is specified, Matplotlib will automatically cycle through a set of default colors for multiple lines.

Similarly the style of the line can also be adjusted by specifying the `linestyle` argument inside the `plt.plot` function

```python
plt.plot(x, x - 0, color='green', linestyle='solid')
plt.plot(x, x - 1, color='green', linestyle='dashed')
plt.plot(x, x - 2, color='green', linestyle='dashdot')
plt.plot(x, x - 3, color='green', linestyle='dotted')

# For short, you can use the following codes:
plt.plot(x, x - 4, color='red', linestyle='-')  # solid
plt.plot(x, x - 5, color='red', linestyle='--') # dashed
plt.plot(x, x - 6, color='red', linestyle='-.') # dashdot
plt.plot(x, x - 7, color='red', linestyle=':')  # dotted
```

To be really consise, both these arguments combined into a single non-keyword arguent within the `plt.plot()` function as follows. 

```python
plt.plot(x, x - 0, '-g')  # solid green
plt.plot(x, x - 1, '--c') # dashed cyan
plt.plot(x, x - 2, '-.k') # dashdot black
plt.plot(x, x - 3, ':r');  # dotted red
```

## Adjusting the Plot: Axes Limits

Most of the time Matplotlib chooses convenient plot limits for your graphs, but for the times where it's necessary to be specific about these limits, plt.xlim() and plt.ylim() methods come handy

```python
plt.plot(x,np.sin(x))
plt.xlim(-2,11)
plt.ylim(-.5,1.1)
```

Say for some reason you would like to reverse the axis, even that can be done by reversing the order of the arguments

```python
plt.plot(x,np.sin(x))
plt.xlim(-2,11)
plt.ylim(1.1,-.5,)
```

Another concise method is the `plt.axis()` method which let's you set the x and y axes limits with just one line of code. Take care of the potential confusion that might arise between the `plt.axis()` and the `plt.axes()` function

```python
plt.plot(x,np.sin(x))
plt.axis([-1,10,-1.1,1.1])
```

This `plt.axis()` method has many more functions like constricting the bounds of the axis limits to contain just the main graph and minimal empty space.

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

It allows even higher-level specifications, such as ensuring an equal aspect ratio so that on your screen, one unit in x is equal to one unit in y

```python
plt.plot(x, np.sin(x))
plt.axis('equal')
```

## Labeling Plots

Let's now look at labeling of plots: titles, axis labeling and simple legends.

```python
plt.plot(x,np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('sin(x)')
```

The size and style of these labels can be adjusted by simply looking into the arguments of the methods above.

While showing multiple lines in a single axes, it can be usefull to label each of these lines and display a legend that shows what each line stands for. Although there are multiple ways of doing this, the most preferable way is the following one.

```python
plt.plot(x,np.sin(x),'--r',label='Sin(x)')
plt.plot(x,np.cos(x),label='Cos(x)')
plt.legend()
```
