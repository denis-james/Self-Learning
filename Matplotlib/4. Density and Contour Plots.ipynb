# Density and Contour Plots

Sometimes visualizing 3d plots can be done using 2d plots with color coded regions that show variying intensities. This can be done using three primary functios. `plt.contour` for contour plots, `plt.contourf` for filled contour plots, and plt.`imshow` for showing images.

## Visializing 3D function

Let's first beging with the simple format of 3dimensional function $z=f(x,y)$

````python
def f(x,y):
    return np.sin(x)*np.sin(x*y)+3*np.sin(x)**3
````

A contour plot takes two primary arguments:
- a grid of x values
- a grid of y values
- a grid of z values

The x and y values are positions on the x-y plane and the z value represents the intensity of each of these x-y combination.
For this we use the `np.meshgrid` function that computes a two dimensional grid from one dimensional grid.

````python
x=np.linspace(-1,5,100)
y=np.linspace(-1,5,100)
X,Y=np.meshgrid(x,y)
f(X,Y)
````

````python
plt.contour(X,Y,f(X,Y),colors='black')
````

Notice how the negative values have been dashed and the positive values are being represented by solid lines. This can be altered and a color scheme can be set as such. 

Also the density of the lines can also be altered by manually specifying how many lines need to be filled within each consecutive line 

````python
plt.contour(X,Y,f(X,Y),30,cmap='RdGy') # RdGy being short for Red-Gray
````

The multiple values of colormap that Matplotlib allows can be accessed as follows.

```python
plt.cm.
```

The plot looks quite empty. To solve this we can use the filled contour plot function  `plt.contourf()` which works the same way as `plt.coutour()`.

````python
plt.contourf(X,Y,f(X,Y),32,cmap='RdGy')
plt.colorbar()
````

The colorbar shows where the graph is positive and negative.

One problem that arises now is the plot being too "jittery", meaning the colored regions seeming discrete and not continuous and smooth. One way this can be solved is by setting the contour to a really high number. But this tends to hurd the efficiency of the plots. 

This is where the `plt.imshow()` function comes into play.

````python
plt.imshow(f(X,Y),extent=[-1,5,-1,5],cmap='RdGy',origin='lower')
plt.colorbar()
````

A few points to keep in mind about this function is that:
- This function plots the points regardless of the range of the x values or the y values given into the function, which is clearly evident from how it doesn't take any such arguments
- It basically plots a grid of values and smoothens it out, so the domain of values have to be labeled on the plot for it to make sense. This label, as they are set manually, can be set to whatever value you want irrespective of what the inputs of the functions have been to generate the grid of z values.
- By default, cause of it being an image visializing function, the origin is set to be the top left corner, which is not the case while plotting graphs. Thus this parameter `origin` has to be manually set to `lower`
- The function sets the aspect ratio of the image to the aspect ratio of the input data. This can be changed using the `axpect` argument to whichever style prefered

The two functions studied earlier can also be combined into the same plot to show a better picture of the data we're trying to plot.

````python
ContourPlot=plt.contour(X,Y,f(X,Y),colors='black')
plt.clabel(ContourPlot)
plt.imshow(f(X,Y),alpha=0.5,origin='lower',extent=[-1,5,-1,5],cmap='GnBu_r')
plt.colorbar()

````
