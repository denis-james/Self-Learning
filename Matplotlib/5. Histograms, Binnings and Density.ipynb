# Histograms, Binnings and Density

A simple Histogram can reveal a lot about the dataset we're studying.

```python
data=np.random.randn(1000)
```

```python
plt.hist(data)
```

The `plt.hist()` function has a lot of customizable arguments that can be tweeked as per requirement.

```python
plt.hist(
    data,
    bins=30,
    alpha=.5,
    edgecolor='black',
    histtype='stepfilled',
    color='red'
)
```

Multiple plots combined with a little bit of transparancy and difference in color turns out to produces some visually pleasing results

```python
data1=np.random.normal(0,5,1000)
data2=np.random.normal(7,5,1000)
data3=np.random.normal(-3,1,1000)

arg=dict(alpha=0.3, bins=40,edgecolor='black',histtype='stepfilled')
plt.hist(data1,**arg,)
plt.hist(data2,**arg)
plt.hist(data3,**arg)
```

Say, in the graph above, you want to know the y values of each x coordinate, you can do so by simply using the `np.histogram()` function

```python
np.histogram(data1,bins=30)
```
