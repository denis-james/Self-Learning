# Histograms, Binnings and Density

A simple Histogram can reveal a lot about the dataset we're studying.

```python
data=np.random.randn(1000)
```

```python
plt.hist(data)
```

The `plt.hist()` function has a lot of customizable arguments that can be tweeked as per requirement.

```python
plt.hist(
    data,
    bins=30,
    alpha=.5,
    edgecolor='black',
    histtype='stepfilled',
    color='red'
)
```

Multiple plots combined with a little bit of transparancy and difference in color turns out to produces some visually pleasing results

```python
data1=np.random.normal(0,5,1000)
data2=np.random.normal(7,5,1000)
data3=np.random.normal(-3,1,1000)

arg=dict(alpha=0.3, bins=40,edgecolor='black',histtype='stepfilled')
plt.hist(data1,**arg,)
plt.hist(data2,**arg)
plt.hist(data3,**arg)
```

Say, in the graph above, you want to know the y values of each x coordinate, you can do so by simply using the `np.histogram()` function

```python
np.histogram(data1,bins=30)
```

## Two-Dimensional Histograms and Binnings

We just now created a one dimensional Histogram. This can also be done in two dimensions with the x-y plane being broken down into multiple segments. 

Let's first generate some 1000 points in 2dimensional plane following the Bivariate Normal Distribution

```python
mean=[0,0]
covarianceMatrix=[
    [1,1],
    [1,2]
]
[x,y]=np.random.multivariate_normal(mean,covarianceMatrix,10000).T
```

### `plt.hist2d`: Two-dimensional histogram

one straightforward way to plot a two-dimensional histogram is to use the function below as such.

```python
plt.cm.
```

```python
plt.hist2d(x,y,bins=60,cmap='OrRd')
plt.colorbar().set_label('Counts')
```

Just like `plt.hist()` had multiple other keword arguments, `plt.hist2d()` also has a plathora of such arguments that can be explored to customize plots.

Also, just like there exists `np.histogram()` for `plt.hist()`, there is `np.histogram2d()` for `plt.hist2d()` which can be used as follows.

```python
(binvalues,xaxisvalues,yaxisvalues)=np.histogram2d(x,y,bins=30)
```

### `plt.hexbin()`: Hexagonal binnings

We can also create hexagonal tessalations instead of the commonly used square grid

```python
plt.hexbin(x, y, gridsize=30, cmap='Blues')
plt.colorbar(label='count in bin')
```
