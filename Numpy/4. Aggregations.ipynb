# Aggregations: Min, Max, and Everything In Between

Sum of Numpy Array

```python
np.random.seed(39)
L=np.random.random(100)
print(sum(L))#Python's function (Slower)
print(np.sum(L))#Numpy's Function (Faster)
```

Minimum and Maximum

```python
print(min(L),np.min(L))
print(max(L),np.max(L))
#Similar Implementation only differing in Computation speed and complexity
```

```python
#Also works like this
L.min(),L.max()
```

Multidimnesional Aggregates

```python
R10x10list=np.random.randint(1,100,size=(10,10,10))
print(R10x10list)
R10x10list.sum()#Sum of all the integers in the list
```

```python
R10x10list.min(axis=2)#Asix standing for the deapth in which the min should be calculated. 
#Axis 2 in a 3 dimensional Array meaning min being calculated in the direction of the deepest stack of numbers
```

## Example

```python
import pandas as pd
Data=pd.read_csv('president_heights.csv')
Heights=np.array(Data['height(cm)'])
print(Heights)
```

```python
print("mean Height = ",Heights.mean())
print("standard Deviation = ",Heights.std())
print("minimum Height = ",Heights.min())
```

```python
print("25th percentile = ", np.percentile(Heights,25))
print("50th percentile = ", np.percentile(Heights,50))
print("median = ", np.median(Heights))
print("75th percentile = ", np.percentile(Heights,75))
```
