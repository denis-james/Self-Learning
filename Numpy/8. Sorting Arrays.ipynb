# Array Sorting

Althought Python has it's own `sort` and `sorted` function to work with lists, We'll stick to studying `np.sort` as it turns out to be much more efficient. The `np.sort` function uses quicksort as a default sorting algorithm

```python
x=randi(100,size=10)
```

```python
np.sort(x)#RETURNS THE SORTED ARRAY. DOES NOT CHANGE THE ORIGINAL ARRAY
```

```python
x.sort()#SORTS THE ARRAY IN PLACE
```

```python
np.random.shuffle(x)
x
```

```python
sortedindeces=np.argsort(x)#argsort RETURNS incedes of the sorted arrays
```

```python
x[sortedindeces]
```

## Sorting along rows or columns(or any specific axis)

```python
np.random.seed(11)
X3D=np.arange(1000)
np.random.shuffle(X3D)
X3D=X3D.reshape((10,10,10))
X3D
```

```python
np.sort(X3D,axis=2)
```

## Partitioning

`np.partition` function takes the array `A` and an integer `K` as an argument and returns an array with `K` smallest values of `A` on the left side of the array and the rest values on the right side of it, in an arbitrary order

```python
x,X3D
```

```python
np.partition(x,4)
```

```python
np.partition(X3D,4,axis=2)
```

Notice how along axis 2 (the deepest axis) all the first three elements are the smallest out of all the other elements in the list.
